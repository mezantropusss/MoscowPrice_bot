```python
# Выводим уникальные значения в столбце 'Renovation'
unique_renovation = data['Renovation'].unique()

print(unique_renovation)
```

```python
#Создание функции для замены значений в столбце Renovation
def rename(renovation):
    if renovation == 'Cosmetic':
        return 0
    elif renovation == 'European-style renovation':
        return 1
    elif renovation == 'Without renovation':
        return 2
    elif renovation == 'Designer':
        return 3
    else:
        return None
```

```python
#Замена значений в столбце Renovation на цифры
data['Renovation'] = data['Renovation'].apply(rename)
data['Renovation'] 
```

```python
# Создаем маску для квартир, расположенных на последнем этаже
last_floor_mask = data['Floor'] == data['Number of floors']

# Выводим все квартиры, расположенные на последнем этаже
last_floor_apartments = data[last_floor_mask]

print(last_floor_apartments)
```

```python
# Создаем маску для квартир, расположенных на последнем этаже
last_floor_mask = data['Floor'] == data['Number of floors']

# Создаем два массива: стоимость квартир на последнем этаже и стоимость квартир на других этажах
last_floor_prices = data[last_floor_mask]['Price']
other_floor_prices = data[~last_floor_mask]['Price']

# Выполняем t-тест для независимых выборок
t_statistic, p_value = stats.ttest_ind(last_floor_prices, other_floor_prices)

print('t-статистика:', t_statistic)
print('p-значение:', p_value)
```
